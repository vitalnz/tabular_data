```python
# Importing libraries
import pandas as pd
from IPython.display import display, HTML
```

```python
# Read data
t = pd.read_csv(r'C:\training2024\tracers.csv')
```

```python
# Count the total population size
tot = int(t.shape[0])
# Print the population size
print("The total population size before selection is {}".format(tot))
```

```python
# Set percentage to be used during random selection
perc = 0.35
# set sample size to be selected
sample_size = tot*perc
```

```python
# Random selection using sample_size
selected = t.sample(n=int(sample_size))
```

```python
print('The percentage used for random selection is {}'.format(perc))
```

```python
selected_html = selected.to_html(index=False)
print('Selected tracers:', selected.shape[0])
display(HTML(selected_html))
```

```python
# Export selected sample
selected.to_csv(r"C:\training2024\selected_tracers.csv", index=False)
```
